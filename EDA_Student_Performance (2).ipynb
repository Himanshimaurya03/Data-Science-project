# Exploratory Data Analysis (EDA)
## Dataset: Student Performance Dataset


## Objective

The objective of this project is to perform Exploratory Data Analysis (EDA) on the Student Performance dataset. 
The goal is to understand the data, identify patterns, analyze relationships between variables, and gain insights 
into factors affecting student performance using statistical methods and visualizations.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('StudentsPerformance.csv')
df.head()
```

```python
df.shape

df.columns

df.info()
```

### Dataset contains student details like gender, study time, and marks.
It includes both categorical and numerical data.

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.duplicated().sum()

df = df.drop_duplicates()
```

```python
df.describe()
```

### Observations:
- Average marks are in mid range
- There are variations in scores among students

```python
df['gender'].value_counts().plot(kind='bar')
plt.title('Gender Distribution')
plt.show()
```

```python
df['math score'].plot(kind='hist')
plt.title("Math Score Distribution")
plt.show()
```

```python
df['reading score'].plot(kind='hist')
plt.title("Reading Score Distribution")
plt.show()
```

```python
df.groupby('gender')['math score'].mean().plot(kind='bar')
plt.title('Average Math Score by Gender')
plt.show()
```

```python
df['gender'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Gender Distribution (Pie Chart)")
plt.ylabel("")  # removes extra label
plt.show()
```

```python
import seaborn as sns

corr = df[['math score','reading score','writing score']].corr()

sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```
